```julia
function simulateprocess(x1; a = 1.0, b = 1.0, N = 5, d::Sampleable{Univariate,Continuous})
    x = zeros(typeof(x1), N+1) # preallocate vector, careful on the type
    x[1] = x1
        for t in 2:N+1
        x[t] = a * x[t-1] + b * rand(d) # draw
        end
    return x
end
```

```julia
@show simulateprocess(0.0, d=Normal(.2, 2.))
```
